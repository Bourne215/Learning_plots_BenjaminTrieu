```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

spotify_df = pd.read_csv(r"../data/Cleaned_Spotify_2024_Global_Streaming_Data.csv")
spotify_df.head()
```

Bar Plot Exercise

```python
#group by genre and calculate the average total streams
avg_streams = spotify_df.groupby("Genre")["Total Streams (Millions)"].mean().sort_values(ascending=False)
avg_streams
```

```python
plt.bar(avg_streams.index, avg_streams.values)
plt.show()
```

```python
plt.bar(avg_streams.index, avg_streams.values)

plt.xlabel("Genre")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")

plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(avg_streams.index, avg_streams.values)

plt.xlabel("Genre")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(avg_streams.index, avg_streams.values)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel("Genre")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
avg_artist_streams = spotify_df.groupby("Artist")["Total Streams (Millions)"].mean().sort_values(ascending=False)
avg_artist_streams
```

```python
plt.Figure(figsize=(10,6))
colors = ["#60ac9b"]*len(avg_artist_streams)

max_index = avg_artist_streams.idxmax()
highlight_position = avg_artist_streams.index.get_loc(max_index)
colors[highlight_position]= "#bba82a"

plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width=0.8)
#Side note: I'm not doing pink.
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel("Artist")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Artist")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))

special_artist = "Ariana Grande"
colors = ["#33107e" if artist == special_artist else "gray" for artist in avg_artist_streams.index]

plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width=0.9)
#Side note: I'm not doing pink.
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel("Artist")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Ariana Grande Ranks 2nd in Total Average Streams")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

Scatter Plot Exercise

```python
spotify_df.columns
spotify_df.info()
```

```python
plt.scatter(spotify_df["Streams Last 30 Days (Millions)"], spotify_df["Total Streams (Millions)"])
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.scatter(
    spotify_df["Streams Last 30 Days (Millions)"], 
    spotify_df["Total Streams (Millions)"],
    color = "#078b0c",
    alpha = 0.8,
    s = 25)

plt.xlabel("Streams this month")
plt.ylabel("Total Streams")
plt.title("No Apparent Correlation Between Monthly Listens and Total Streams")

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))

feature_genre = "Pop"

colors = ["#056508" if genre == feature_genre else "#8fae92" for genre in spotify_df["Genre"]]

plt.scatter(
    spotify_df["Streams Last 30 Days (Millions)"], 
    spotify_df["Total Streams (Millions)"],
    color = colors,
    alpha = 0.8,
    s = 25)

plt.xlabel("Streams this month")
plt.ylabel("Total Streams")


ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.set_title("Do Pop Songs Achieve Higher Monthly Listens or Total Streams?", fontsize = 18, pad = 35)

#I know you set_alpha to 0.5
#but the lines are  non-existent if I set it that way so I had to set it to 1
for spine in ax.spines.values():
    spine.set_linewidth(0.25)
    spine.set_alpha(1)


plt.tight_layout()
plt.show()
```

Line Charts Exercise

```python
yearly_avg = spotify_df.groupby("Release Year")["Total Streams (Millions)"].mean().round(2)
yearly_avg
```

```python
plt.figure(figsize= (12,6))
plt.plot(yearly_avg.index,
         yearly_avg.values,
         marker= "o",
         linestyle = "--",
         color = "#08c514",
         linewidth = 0.75)

plt.xlabel("Release Year")
plt.ylabel("Average Total Streams")
plt.title("Did Listeners Stream Less in 2022?", fontsize = 18, pad = 20)

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.axvline(
    2022,
    color = "#438a08",
    linewidth = 0.75,
    alpha = 0.7)

ax.text(
    2022+0.05,
    2225,
    "Dip In Total Steams",
    color = "#438a08",
    fontsize = 8
)

plt.ylim(0, 3500)

plt.show()
```

Pie Chart Exercise

```python
category = "Platform Type"
values = "Total Streams (Millions)"

platform_totals = spotify_df.groupby(category)[values].sum()
platform_totals
```

```python
plt.figure(figsize=(8, 10))
plt.pie(
    platform_totals,
    labels=None,
    startangle=110,
    colors=["gray","#047d0c"],
    autopct="%1.1f%%")

plt.title("Near Even Split Between Premium and Free Users")

plt.legend(
    labels = platform_totals.index,
    loc="lower right",
    title = "Platform"
)

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))

wedges, texts, autotexts = ax.pie(
    platform_totals,
    autopct="%1.1f%%",
    startangle=110,
    colors=["gray","#047d0c"],
    )

ax.set_title(
    "Near Even Split Between Premium and Free Users",
    fontsize = 20,
    pad = 15,
    loc = "center"
)

for autotext in autotexts:
    autotext.set_fontsize(23)
    autotext.set_color("white")
    autotext.set_weight("bold")

ax.legend(
    wedges,
    platform_totals.index,
    loc="lower center",
    bbox_to_anchor = (0.5, 0.9),
    ncol = 2
)

plt.tight_layout()
plt.show()
```

Histograms Exercise

```python
plt.figure(figsize=(10, 6))
plt.hist(
    spotify_df["Avg Stream Duration (Min)"],
    bins=15,
    color = "#037a17",
    alpha = 0.9,
    edgecolor = "white"
)

plt.xlabel("Average Stream Duration (Minutes)")
plt.ylabel("Number of Tracks")
plt.title("Distibution of Average Stream Duration", fontsize = 16, pad = 20)

ax = plt.gca()
for spine in ax.spines.values():
    spine.set_linewidth(0.5)
    spine.set_alpha(0.5)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)



plt.show()
```

Heatmaps Exercise

```python
numeric_spotify_df = spotify_df.select_dtypes(include=[np.number])
corr = numeric_spotify_df.corr()

im = plt.imshow(corr, vmin=-1, vmax=1, cmap="coolwarm")
plt.colorbar(im, label="Pearson Corellation")

plt.xticks(range(len(corr.columns)), corr.columns, rotation = 45, ha="right")
plt.yticks(range(len(corr.columns)), corr.columns)

plt.show()
```
